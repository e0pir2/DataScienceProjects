# Get to know the data

### Missing data

```python
print("Number of missing data in df_train: {}".format(df_train.isnull().any().sum()))
print("Number of missing data in df_test: {}".format(df_test.isnull().any().sum()))
```

### Univariate Data

```python
for col in df_train.columns[2:]:
    fig,sub=plt.subplots(1,1,figsize=(16,8))
    sns.distplot(df_train[col],ax=sub,hist_kws={"edgecolor":"black","alpha":0.3},color="blue")
    sns.distplot(df_test[col],ax=sub,hist_kws={"edgecolor":"black","alpha":0.3},color="red")
    sub.grid()
```
