```python
import requests
from bs4 import BeautifulSoup
import json
from datetime import datetime
import numpy as np
import pandas as pd
import time
import re
from tqdm import tqdm
```

```python
link_part1="https://www.sueddeutsche.de/news/page/"
link_part2="?search=%22depressionen%22&sort=date&dep%5B%5D=sport&dep%5B%5D=wissen&dep%5B%5D=gesundheit&all%5B%5D=typ&all%5B%5D=sys&all%5B%5D=time"
```

```python
json_list=[]
for i in range(1,33,1):
    current_page=link_part1+str(i)+link_part2
    print("new Link")
    print("\n")
    print(current_page)
    page=requests.get(current_page)
    page_content=BeautifulSoup(page.content,"html.parser")

    attribute=["Artikel_HL","Ressort","Veroeffentl","Kurzbeschr","Quelle","Link"]
    
    for artikel in page_content.find_all("div",class_="entrylist__entry"):
        
        json_file={k:np.nan for k in attribute}
        try:
            json_file["Link"]=artikel.find_all("a",{"class":"entrylist__link","href" : re.compile('https://www.sueddeutsche.de/*')})[0]["href"]
        except:
            pass
        try:
            json_file["Ressort"]=artikel.find_all("a",{"class":"entrylist__link","href" : re.compile('https://www.sueddeutsche.de/*')})[0].find_all("strong")[0].text
        except:
            pass
        try:
            json_file["Artikel_HL"]=artikel.find_all("a",{"class":"entrylist__link","href" : re.compile('https://www.sueddeutsche.de/*')})[0].find_all("em")[0].text
        except:
            pass
        print(json_file["Artikel_HL"])
        try:
            json_file["Quelle"]=artikel.find_all("span")[0].text
        except:
            pass
        json_file["Kurzbeschr"]=artikel.find_all("p")[0].text.replace("\n","")[86:]
        json_file["Veroeffentl"]=artikel.find_all("time")[0].text.replace("\n","")[4:]

        json_list.append(json_file)
```

```python
df=pd.DataFrame(json_list)
```

```python
df["Veroeffentl"]=df.apply(lambda x: x["Veroeffentl"][:10],axis=1)
```
